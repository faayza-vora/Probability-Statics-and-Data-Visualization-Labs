```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
# function to find frequency dustribution
def hist_plot(img):

    hist =[0]*256  # empty list to store the count of each intensity value
    l,b=img.shape
    for i in range(l):
        for j in range(b):
            hist[img[i,j]]+=1

    return list(range(256)),hist
```

```python
def noisy_noise(img, mu, sigma2):
    x, y = img.shape
    sigma = np.sqrt(sigma2)
    n = np.random.normal(loc = mu, scale = sigma, size = (x, y))
    nf = np.uint8(n)
    return nf
```

```python
def otsu_binarise(img):
    bins,hist = hist_plot(img)
    # Getting the ftrequency for different intensities
    l,b = img.shape
    # Total number of pixels to find the probability
    total = l*b

    max_var = 0
    pivot = 0 # Value at which the background and foreground pixels are seperated


    # Computing between-class variance

    for t in range(0,256):

        # class probabilities
        w0 = np.sum(hist[:t]) / total
        w1 = np.sum(hist[t:]) / total

       # class means
        mu0 = 0
        for k in range(0,t):
            mu0 += k * hist[k]  # finding sum
        mu0 /= (np.sum(hist[:t]) + 1e-5)  # Normalization
        mu1 = 0
        for l in range(t,256):
           mu1 += l * hist[l]  # findin sum
        mu1 /= (np.sum(hist[t:]) + 1e-5)  # Normalization

        # Compute between class variance
        var = w0 * w1 * (mu0 - mu1) ** 2

        # Finding maximum variance
        if var > max_var:
            max_var = var
            pivot = t

    # New image initialization
    final = np.zeros_like(img , dtype = 'uint8')
    final[img > pivot] = 255

    # print(pivot)
    return final
```

```python
# Inputs

# @title
#@markdown Enter mean value for the noise
mu_0 = 0 # @param {type:"number"}
#@markdown Enter the variance for noise
sigma2_0 = 30 # @param {type:"number"}

# @title
#@markdown Enter another mean value for the noise
mu_1= 0 # @param {type:"number"}
#@markdown Enter another variance for noise
sigma2_1 = 60 # @param {type:"number"}

# @title
#@markdown Enter another mean value for the noise
mu_2 = 0 # @param {type:"number"}
#@markdown Enter another variance for noise
sigma2_2 = 100 # @param {type:"number"}

img = cv2.imread('/content/drive/MyDrive/image processing/bookpage_1.jpeg', 0)

# Calling functions

final = otsu_binarise(img)
n0 = noisy_noise(img, mu_0, sigma2_0)
n1 = noisy_noise(img, mu_1, sigma2_1)
n2 = noisy_noise(img, mu_2, sigma2_2)
# Showing images


# sources with noise

new_source_0 = img + n0
new_source_1 = img + n1
new_source_2 = img + n2

# binarizing noisy images

final0 = otsu_binarise(new_source_0)
final1 = otsu_binarise(new_source_1)
final2 = otsu_binarise(new_source_2)

# showing images

fig = plt.figure(figsize=(30, 35))

# Orignal image
fig.add_subplot(4, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Source image")
plt.axis('off')

# binarized
fig.add_subplot(4, 3, 2)
plt.imshow(cv2.cvtColor(final, cv2.COLOR_BGR2RGB))
plt.title("Binarized")
plt.axis('off')
# empty
fig.add_subplot(4, 3, 3)
plt.axis('off')
#noise1
fig.add_subplot(4, 3, 4)
plt.imshow(cv2.cvtColor(n0, cv2.COLOR_BGR2RGB))
plt.title("Low Noise")
plt.axis('off')
# new source 1
fig.add_subplot(4, 3, 5)
plt.imshow(cv2.cvtColor(new_source_0, cv2.COLOR_BGR2RGB))
plt.title("New Source")
plt.axis('off')

# binarized
fig.add_subplot(4, 3, 6)
plt.imshow(cv2.cvtColor(final0, cv2.COLOR_BGR2RGB))
plt.title("Binarized")
plt.axis('off')

# Noise 2
fig.add_subplot(4, 3, 7)
plt.imshow(cv2.cvtColor(n1, cv2.COLOR_BGR2RGB))
plt.title("Medium Noise")
plt.axis('off')

# new source
fig.add_subplot(4, 3, 8)
plt.imshow(cv2.cvtColor(new_source_1, cv2.COLOR_BGR2RGB))
plt.title("New Source")
plt.axis('off')

# Binarized
fig.add_subplot(4, 3, 9)
plt.imshow(cv2.cvtColor(final1, cv2.COLOR_BGR2RGB))
plt.title("Binarized")
plt.axis('off')

# Noise 3
fig.add_subplot(4, 3, 10)
plt.imshow(cv2.cvtColor(n2, cv2.COLOR_BGR2RGB))
plt.title("High Noise")
plt.axis('off')

# new source
fig.add_subplot(4, 3, 11)
plt.imshow(cv2.cvtColor(new_source_2, cv2.COLOR_BGR2RGB))
plt.title("New Source")
plt.axis('off')

# Binarized
fig.add_subplot(4, 3, 12)
plt.imshow(cv2.cvtColor(final2, cv2.COLOR_BGR2RGB))
plt.title("Binarized")
plt.axis('off')

plt.show()




```

It is a widely used method in image processing for automatic thresholding. It aims to seperate an image into two classes named foreground and background by finding an optimal threshold value based on the histogram pixel intensities.

##Image thresholding
It refers to a technique which is used to seperate region of interest from background by converting it into a binary image. This is done by using otsu's binarizatio as stated above

##Gaussian noise
It is a type of noise which follows gaissian or normal distribution. It is commonly encountered in digital images and can degrade image quality. In order to generate gaussian noise i have used np.random.normal function

##Between-class Variance
Between-class variance is a measure of the spread between the mean values of different classes in an image histogram. It quantifies the separation between foreground and background classes and is maximized by Otsu's Binarization to find the optimal threshold.

##In-class Variance
In-class variance measures the spread of pixel intensities within each class (foreground and background). Minimizing in-class variance ensures that each class is internally homogeneous after thresholding which in turn ensures optimal calculation of threshold.

##Formula
Otsu's Method: Otsu's method calculates the optimal threshold value
threshold
threshold by maximizing the between-class variance $\sigma_{b}^{2}$  is defined as:
\begin{equation}
     \sigma_{b}^{2} = w_{0} \times w_{1} \times (\mu_{0} - \mu_{1}) ^{2}
\end{equation}
where $w_{0}$ and $w_{1}$ are the probabilities of the two classes (foreground and background) separated by the threshold and $\mu_{0}$ and $\mu_{1}$ are the mean intensities of the two classes.
