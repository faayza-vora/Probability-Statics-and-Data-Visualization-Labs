###Histogram equalization
It is a technique used in image processing to improve contrast by redistributing the intensity values of an image's histogram. This process aims to make the histogram more uniformly distributed across the intensity range.
In order to achieve the function, hist_plot iterates through the image and calculates the frequencies for all the intensity values. This is then used to calculate PMF. CDF is then calculated by using np.cumsum. cumsum is used because we are dealing with intensity of different pixels which is discrete. Now, to make the distribution uniform, a new array is created, which includes the CDF's values multiplied by 255. This gave us equalized CDF, which was used to get equalized images for both the target and source.
###Histogram matching
It is a method used to change an image's histogram to match a specified histogram. This is often used to adjust the contrast of an image to make it similar to a reference image to get the desired histogram distribution.
In order to achieve this, the index of the nearest CDF in the target image is compared with the index of the source image. This is computed by the function min_in. The maximum index is used if the deviation is the same for multiple indices. These indices are then added in a final 2D array, which is then plotted as a matched image and the matched histogram.
