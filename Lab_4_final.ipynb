2. Plot the Cumulative Distribution Function (CDF) of each continuous random variable above by varying parameters values taken as user input if the closed form mathematical expression exists. [1 Mark]

a) Uniform

```python
# @markdown enter a pair of a,b <br>
a1 = 5 # @param {type:"number"}
b1 = 7 # @param {type:"number"}
print("Q2a")
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(a1,b1,10000)
y=[]
y3=[]
for k in range(len(x)):
  y.append(1/(abs(b1-a1)))
  y3.append(abs((x[k]-a1)/(b1-a1)))
y1=np.cumsum(y)
y2=y1/y1[-1]
ax=plt.axes()
plt.plot(x,y3,alpha=0.6,label='formula')
plt.plot(x,y2,"--",label='cumsum')
plt.xlabel("Outcomes")
plt.ylabel("CMF")
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
```

3. Plot the CDF of each discrete random variable given in Lab Assignment 3 by varying parameters values taken as user input (No need to derive closed form expression in this case). [1 Mark]

 a) Discrete Uniform

```python
# @title
# @markdown enter the starting value
l = 2 # @param {type:"integer"}
# @markdown enter the ending value
h = 7 # @param {type:"integer"}

import numpy as np
import matplotlib.pyplot as plt
print("Q1(a)")
plt.figure(facecolor="#43766c")
print()
x=np.arange(l,h)
y=np.ones_like(x)/len(x)
y1=np.cumsum(y)/np.sum(y)
ax=plt.axes()
ax.set_facecolor("#43766c")
plt.step(x,y1,color="#561c24")
plt.title(f'DISCRETE UNIFORM CDF FOR ({l},{h})')
plt.xlabel("Outcomes")
plt.ylabel("CMF")
plt.show()
```

b) Bernoulli

```python
# @title
# @markdown enter the probability of success
p = 0.7 # @param {type:"number"}

import numpy as np
import matplotlib.pyplot as plt
def cdf(a,p):
    if a<0:
        return 0
    if a>0 and a<1:
        return p
    if a==0:
        return 1-p
    return 1

print("Q1(b)")
print()
x=[-3,-2,-1,0,1,2,3]
y=[1-p,p]
y1=[]
for i in x:
    y1.append(cdf(i,p))
plt.figure(facecolor="white")
ax=plt.axes()
ax.set_facecolor("white")
plt.step(x,y1,color="black")
plt.title(f'BERNOULLI CDF FOR ({p},{round(1-p,1)})',color="black")
plt.xlabel("Outcomes",color="black")
plt.ylabel("CMF",color="black")
plt.show()
```

c) Binomial

```python
# @title
# @markdown enter the probability of success
p = 0.5 # @param {type:"number"}
# @markdown enter the number of trials
n = 7 # @param {type:"number"}
import numpy as np
import matplotlib.pyplot as plt
def fac(n):  #function to find factorial
  if n == 0:
    return 1
  p=1
  for j in range(1,int(n+1)):
    p*=j
  return p
print("Q1(c)")
print()
x=np.arange(0,n+1) # since possible valuse of successes can be 0 to n
y=[]
for k in x:
  y.append(((fac(n)/(fac(k)*fac(n-k)))*(p**k)*((1-p)**(n-k))))
y1=np.cumsum(y)/np.sum(y)
plt.figure(facecolor="#f2afef")
ax=plt.axes()
ax.set_facecolor("#f2afef")
plt.step(x,y1,color="#33186b")
plt.title(f'BINOMIAL CDF FOR ({n=},{p=})')
plt.xlabel("Outcomes")
plt.ylabel("CMF")
plt.show()
print()
```

d) Poisson

```python
# @title
# @markdown enter the average rate of success
lam = 10 # @param {type:"number"}
# @markdown enter the number of trials
n = 20 # @param {type:"number"}
import numpy as np
import matplotlib.pyplot as plt

def fac(n): # Function to calculate factorial
    p=1
    if n == 0:
        return 1
    else:
      for i in range(1,n+1):
        p*=i
    return p

print("Q1(d)")
print()
x = np.arange(0, n+1) # Possible values of k
y = []
plt.figure(facecolor="#000000")
for k in x:
    pmf = np.exp(-lam) * (lam**k) / fac(k)
    y.append(pmf)
y1=np.cumsum(y)/np.sum(y)
ax=plt.axes()
ax.set_facecolor("#000000")
ax.spines['bottom'].set_color('white')
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')
ax.spines['left'].set_color('white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
plt.step(x, y1,color="#f80164")
plt.title(f'Poisson CDF for $\lambda$={lam}',color="white")
plt.xlabel('Number of Occurrences (k)',color="white")
plt.ylabel('CMF',color="white")
plt.show()

print()
```
