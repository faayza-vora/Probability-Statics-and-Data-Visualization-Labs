#### Question 4. How does the height of the player influence their performance? Find its correlation coefficient and give a scatter plot.

##### Finding Correlation coefficient

```python
def mean(c):
    m = np.sum(c) / len(c)
    return m
def var(c):
    m = mean(c)
    var = (np.sum((c - m) ** 2)) / (len(c) - 1)
    return var
def cr(c_1 , c_2):
    m_1 = mean(c_1)
    s_1 = np.sqrt(var(c_1))
    m_2 = mean(c_2)
    s_2 = np.sqrt(var(c_2))
    x = (c_1 - m_1) / s_1
    y = (c_2 - m_2) / s_2
    pr = x * y
    sm = np.sum(pr)
    res = sm / (len(c_1) - 1)
    return res
```

Answer

```python
# getting heights of players from players dataset and No.of wins from the merged years dataset.
dfm_new = dfm.groupby('Winner').count()
Winners = dfm_new.index.to_numpy()
Wins = dfm_new['ATP'].to_numpy()
new = pd.DataFrame(Winners,Wins)
new.reset_index(inplace = True)
dfp_new = dfp[['full_name','height_cm']]
final = pd.merge(new , dfp_new , how = 'outer' , left_on = 0 , right_on = 'full_name').dropna().drop_duplicates().drop(columns=[0])
final = final.rename(columns={'index': 'Wins'})
# Findin correlation coefficient between wins and height
a_0  = final.iloc[:, 0].to_numpy()
a_1  = final.iloc[:, 2].to_numpy()
a_0 = list(a_0)
a_1 = list(a_1)
zero_idx = a_1.index(0)
a_0.pop(zero_idx)
a_1.pop(zero_idx)
zero_idx = a_1.index(0)
a_0.pop(zero_idx)
a_1.pop(zero_idx)
res = cr(a_1 , a_0)
print(f'The correlation coefficient for height and matches won is {res}')
print()
plt.figure()
plt.scatter(a_1 , a_0)
plt.xlabel('Height (in cm)')
plt.ylabel("No of matches won")
plt.show()
```

The correlation coefficient for height and the matches won is 0.0882 which shows their is not a strong linear relationship between these two parameters. This is also seen in the scatter plot given above.
