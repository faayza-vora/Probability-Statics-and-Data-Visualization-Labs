1. Plot the Probability Mass Functions of the following Discrete Random Variables by varying the appropriate parameters taken as input. Compare the plots of the same PMF plotted with different parameters. [2 Marks]

a) Discrete Uniform


```python
# @title
# @markdown enter the starting value
l = 2 # @param {type:"integer"}
# @markdown enter the ending value
h = 7 # @param {type:"integer"}

import numpy as np
import matplotlib.pyplot as plt
print("Q1(a)")
plt.figure(facecolor="#43766c")
print()
x=np.arange(l,h)
y=np.ones_like(x)/len(x)
ax=plt.axes()
ax.set_facecolor("#43766c")
plt.bar(x,y,width=0.5,color="#561c24",edgecolor="black")
plt.title(f'DISCRETE UNIFORM PMF FOR ({l},{h})')
plt.xlabel("Outcomes")
plt.ylabel("Probability")
plt.show()
print()
print("This is the probability mass function for a discrete uniform distribution")

```

b) Bernoulli



```python
# @title
# @markdown enter the probability of success
p = 0.7 # @param {type:"number"}

import numpy as np
import matplotlib.pyplot as plt
print("Q1(b)")
print()
x=[0,1]
y=[1-p,p]
plt.figure(facecolor="white")
ax=plt.axes()
ax.set_facecolor("white")
plt.bar(x,y,color="black",edgecolor="white")
plt.title(f'BERNOULLI PMF FOR ({p},{round(1-p,1)})',color="black")
plt.xlabel("Outcomes",color="black")
plt.ylabel("Probability",color="black")
plt.show()
print()
print("This is the probability mass function for a bernoulli distribution")


```

c) Binomial



```python
# @title
# @markdown enter the probability of success
p = 0.5 # @param {type:"number"}
# @markdown enter the number of trials
n = 7 # @param {type:"number"}
import numpy as np
import matplotlib.pyplot as plt
def fac(n):  #function to find factorial
  if n == 0:
    return 1
  p=1
  for j in range(1,int(n+1)):
    p*=j
  return p
print("Q1(c)")
print()
x=np.arange(0,n+1) # since possible valuse of successes can be 0 to n
y=[]
for k in x:
  y.append(((fac(n)/(fac(k)*fac(n-k)))*(p**k)*((1-p)**(n-k))))

plt.figure(facecolor="#f2afef")
ax=plt.axes()
ax.set_facecolor("#f2afef")
plt.bar(x,y,width=0.5,color="#33186b",edgecolor="black")
plt.title(f'BINOMIAL PMF FOR ({n=},{p=})')
plt.xlabel("Outcomes")
plt.ylabel("Probability")
plt.show()
print()
print("This is the probability mass function for a binomial distribution")




```

d) Poisson


```python
# @title
# @markdown enter the average rate of success
lam = 2 # @param {type:"number"}
# @markdown enter the number of trials
n = 20 # @param {type:"number"}
import numpy as np
import matplotlib.pyplot as plt

def fac(n): # Function to calculate factorial
    p=1
    if n == 0:
        return 1
    else:
      for i in range(1,n+1):
        p*=i
    return p

print("Q1(d)")
print()
x = np.arange(0, n+1) # Possible values of k
y = []
plt.figure(facecolor="#000000")
for k in x:
    pmf = np.exp(-lam) * (lam**k) / fac(k)
    y.append(pmf)
ax=plt.axes()
ax.set_facecolor("#000000")
ax.spines['bottom'].set_color('white')
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')
ax.spines['left'].set_color('white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
plt.bar(x, y, width=0.8,color="#f80164",edgecolor="black")
plt.title(f'Poisson PMF for $\lambda$={lam}',color="white")
plt.xlabel('Number of Occurrences (k)',color="white")
plt.ylabel('Probability',color="white")
plt.show()

print()
print("This is the probability mass function for a Poisson distribution")


```


e) Geometric

```python
# @title
# @markdown enter the average rate of success
p = 0.5 # @param {type:"number"}
# @markdown enter the number of trial in which success is obtained
k = 20 # @param {type:"number"}
import numpy as np
import matplotlib.pyplot as plt
print("Q1(d)")
print()
x=np.arange(0,k+1)
y=(1-p)**(x-1)*p #formula
plt.figure(facecolor="#d0a2f7")
ax=plt.axes()
ax.set_facecolor("#d0a2f7")
plt.bar(x,y,width=0.5,color="#860a35",edgecolor="black")
plt.title(f'GEOMETRIC PMF FOR p={p}')
plt.xlabel("Outcomes")
plt.ylabel("Probability")
plt.show()
print()
print("This is the probability mass function for a geometric distribution")

```
