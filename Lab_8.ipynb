# Function to find Mean of an attribute

```python
def mean(c):
    m = np.sum(c) / len(c)
    return m
```

#Function to find Variance of an attribute

```python
def var(c):
    m = mean(c)
    var = (np.sum((c - m) ** 2)) / (len(c) - 1)
    return var
```

# Function to find correlation coefficient

```python
def cr(c_1 , c_2):
    m_1 = mean(c_1)
    s_1 = np.sqrt(var(c_1))
    m_2 = mean(c_2)
    s_2 = np.sqrt(var(c_2))
    x = (c_1 - m_1) / s_1
    y = (c_2 - m_2) / s_2
    pr = x * y
    sm = np.sum(pr)
    res = sm / (len(c_1) - 1)
    return res
```

# Bubble Sort

```python
def bubble(a,n):
  if n==1:
    return a
  for i in range(0,n-1):
    if a[i][1]<a[i+1][1]:
      a[i],a[i+1] = a[i+1],a[i]
  return bubble(a,n-1)
```

#### Correlation coefficient
It is a measure between $-1$ to $1$ which denotes the strength and direction of relationship between two random variables that is it tells us how similar the measurements are accross two or more random variables.

#### Method
In order to find the correlation coefficient for two random variables (say $x$ and $y$)following steps are followed:
1. Calculate the values of mean $(\bar{x}, \bar{y})$ and standard deviation $(s(x), s(y))$ for both $x$ and $y$ \\
2. Now calculate the standardised values for $x$ and $y$ which are given by
\begin{equation}
 z(x[i]) = \frac{x[i] - \bar{x}}{s(x)}
\end{equation} \\
\begin{equation}
 z(y[i]) = \frac{y[i] - \bar{y}}{s(y)}
\end{equation}
where $i$ represents the $i^{th}$ element in the respective column \\
3. Now multiply the standardized values for every $i^{th}$ element and add all of then together to get the values of sum. Say both the datasets have $n$ elements in it then the sum is given by:

\begin{equation}
 \sum_{i=0}^{n-1} (z(x[i]) \times z(y[i]))
\end{equation}

4. Now divide this sum obtained by $n - 1$ to get the correlation coefficient.

### Interpretation
The correlation coefficient obtained will always lie between $-1$ and $1$ which shows the strength of the relationship between two random variables. The interpretation of the values are as follows:

1. If the corelation coefficient is near $0$ then the relationship between the two variables is considered to be weak that is they are not much related and the scatter plot obtained is widely spaced

2. If the variance is near $1$ then the relationship between the two variavble is strong that is they are closely related and the scatter plot obtained is closely spaced. The positive sign indicates that the relation is positive which means that as one variable increases the other variable also tends to increase. This can been seen in the scatter plot by the treadline which has a positive slope.

3. If the variance is near $-1$ then the relationship between the two variavble is also strong that is they are also closely related and the scatter plot obtained is also closely spaced. The negative sign indicates that the relation is negative which means that as one variable increases the other variable tends to decrease and vice-versa. This can been seen in the scatter plot by the trendline which has a negative slope.
